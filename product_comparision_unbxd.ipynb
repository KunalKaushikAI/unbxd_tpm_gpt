```python
file1_path = "/content/drive/My Drive/product_export/filtered_data.csv"
file2_path = "/content/drive/My Drive/product_export/final.csv"
```

```python
df_caratlane = pd.read_csv(file1_path)
df_unbxd = pd.read_csv(file2_path)
```

```python
merged_df = pd.merge(df_caratlane, df_unbxd, on='sku', how='outer', indicator=True)
```

```python
difference_df = merged_df[merged_df['_merge'] != 'both']
```

```python
difference_df = difference_df.drop(columns='_merge')
```

```python
difference_df = difference_df.drop_duplicates(subset='sku')
```

```python
output_file_path = "/content/drive/My Drive/product_export/difference_no_duplicates.csv"  # Replace with your actual output path
```

```python
difference_df.to_csv(output_file_path, index=False)
```
